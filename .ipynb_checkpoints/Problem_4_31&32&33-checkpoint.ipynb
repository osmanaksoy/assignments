### Problem 31&32&33

The heat equation

\begin{equation}
\frac{\partial T}{\partial t} = \alpha \frac{\partial ^2 T}{\partial x^2}
\end{equation}

governs the time-dependent temperature distribution in a homogeneous constant property solid under conditions where the temperature varies only in one space dimension. Physically, this may be nearly realized in a long thin rod or ver large (infinite ) wall of finite thickness.
  Consider a large wall of thickness L whose initial temperature is given by $ T(t,x) = c sin \pi x/L$. If the faces of the wall continue to be held at $0^o$, then a solution for the temperature at $t > 0$, $0 \leq x \leq L$ is
  
\begin{equation}
T(t,x) = c exp(\frac{-\alpha \pi^2 t}{L^2}) sin \frac{\pi x}{L}
\end{equation}

For this problem let $c=100^o C$, $L = 1 m$, $\alpha = 0.02 m^2/h$. We will consider two explicit methods of soluiton, A. Simple explicit method. Stability requires that $ \alpha \Delta t / (\Delta x^2) \leq \frac{1}{2} $ for this method, B. Alternating direction explicit (ADE) method. This particular version of the ADE method was suggested by Barakat and Clark(1966). In this algorith, the equation for $ p_j ^{n+1} $ can be solved explicitly starting from the boundary at x = 0, whereas the equation for $ q_j^{n+1} $ should be solved starting at the boundary at x = L. There is no stability constraint on the size of the time step for this method. Develop computer programs to solve the problem described above by methods A, B, Crank Nicolson and simple implicit scheme.

```python
import numpy                       
import matplotlib.pyplot as pyplot  
import math
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
from scipy.linalg import solve
```

```python
c = 100.0                       # constant
L = 1.0                         # Thickness of the wall
alpha = 2e-2                    
x = numpy.linspace(0,1,11)
nx = 11
nt = 101
dx = L/(nx-1)
dt = 0.1
r = (alpha*dt)/(dx**2)
```

```python
T_i = c*numpy.sin(3.14*x/L)          # initial conditions
T_exact = numpy.zeros((nt, len(x)))  
T_exact[0,:] = T_i.copy()
```

```python
#Exact solution

for t in range(1,nt):
    T_exact[t,:] = numpy.sin(3.14*x/L)*numpy.exp((-alpha*(3.14)**2*(t/10.))/L**2)*c
```

```python
# Simple Explicit Method

T_exp = c*numpy.sin(3.14*x/L)           # initial conditions
T_nexp = numpy.zeros((nt,len(T_exp))) 
T_nexp[0,:] = T_exp.copy()

for t in range(1,nt):
    T_nexp[t,1:-1] = T_exp[1:-1] + r*(T_exp[2:]-2*T_exp[1:-1]+T_exp[:-2])
    T_nexp[t,0] = T_exp[0]
    T_nexp[t,-1] = T_exp[-1]
    T_exp = T_nexp[t].copy()
```

```python
nt = 101
#Barakat and Clark ADE Method

T_r = c*numpy.sin(3.14*x/L)                  # initial conditions for starting from right
T_l = T_r[::-1].copy()                       # initial conditions for starting from left

P_n = T_r.copy()
Q_n = T_l.copy()

T_ade = numpy.zeros((nt,len(T_l)))
T_ade[:,:] = T_r.copy()


for t in range(1,nt):
    
   
    P_n[1:-1] = ((P_n[:-2] * r) + (T_r[2:] * r) + (T_l[1:-1] * (1-r))) / (1 + r)
    T_r = P_n.copy()
    
       
    Q_n[1:-1] = (T_l[1:-1]*(1-r)/(1+r)) + (Q_n[:-2] + T_l[2:])*(r/(1+r))
    T_l = Q_n.copy()
    
    
    T_ade[t,:] = (Q_n[::-1] + P_n)/2
```

```python
def Matrix(N, R):
    """Generates the Left-Hand side Matrix"""
    
    d = 2*numpy.diag(numpy.ones(N-2)*(1+1./r))
    ud = numpy.diag(numpy.ones(N-3)*-1,1)
    ld = numpy.diag(numpy.ones(N-3)*-1,-1)
    
    A = d+ud+ld
    
    return A
```

```python
def generateRHS(T, R):
    """Generates the Right-Hand Side"""
    
    h = numpy.zeros_like(T)
    
    h = T[2:] + (((1./r)-1) * T[1:-1]*2) + T[:-2]
    h[0] += T[0]
    
    return h
```

```python
def CrankNicolson(T, A, nt, R):
    """Crank Nicolson solver"""
    
    T_x = numpy.zeros((nt,len(T)))
    T_x[:,:]= T.copy()
    for t in range(1,nt):
        Tn = T.copy()
        h = generateRHS(Tn, r)
        T_interior = solve(A,h)
        T[1:-1] = T_interior.copy()
        
        T_x[t,:] = T.copy()
    
    return T_x
```

```python
def MatrixSIC(N, R):
    """Generates matrix for Simple Implicit Method"""
    d = numpy.diag(numpy.ones(N-2)*(2+1./R))
    ud = numpy.diag(numpy.ones(N-3)*-1,1)
    ld = numpy.diag(numpy.ones(N-3)*-1,-1)
    
    A = d + ud + ld
    
    return A
```

```python
def implicitRHS(T, R):
    """Generates Right-Hand Side for Simple Implicit Method"""
    h = numpy.empty_like(T)
    
    h = (1./R)*T[1:-1]
    h[0] += T[0]
    
    return h
```

```python
def simpleimplicit(T, R, A, nt):
    """Solves simple implicit method"""
    T_h = numpy.zeros((nt,len(T)))
    T_h[:,:]= T.copy()
    
    
    for t in range(1,nt):
        Tn = T.copy()
        h = implicitRHS(Tn, R)
        Tinterior = solve(A, h)
        T[1:-1] = Tinterior.copy()
        
        T_h[t,:] = T.copy()
        
    return T_h    
```

```python
A = MatrixSIC(nx, r)
```

```python
Tini = c*numpy.sin(3.14*x/L)
T_h = simpleimplicit(Tini,r,A,nt)
```
