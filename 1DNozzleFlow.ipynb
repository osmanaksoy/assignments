```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
nx = numpy.linspace(0,3,61)
dx = 0.1
dt = 0.0201
gamma = 1.4
nt = 1400
ntd = numpy.linspace(0,nt,nt)

# initial conditions
A = 1 + 2.2*(nx-1.5)**2
rho_n = 1 - 0.3146*nx
T_n = 1 - 0.2314*nx
v_n = (0.1 + 1.09*nx)*T_n**(0.5)

rho_g = numpy.zeros(61)
v_g = numpy.zeros(61)
T_g = numpy.zeros(61)

rho_f = numpy.zeros(61)
v_f = numpy.zeros(61)
T_f = numpy.zeros(61)

rho_h = numpy.zeros(61)
v_h = numpy.zeros(61)
T_h = numpy.zeros(61)

rho = numpy.zeros(nt)
T = numpy.zeros(nt)
Ma_1 = numpy.zeros(nt)
P_1 = numpy.zeros(nt)
```

```python
#Finite Difference Equations

for t in range(nt):
    
    rho_0 = rho_n.copy()
    v_0 = v_n.copy()
    T_0 = T_n.copy()
    
    # 1
    rho_f[:-1] = -rho_0[0:-1]*((v_0[1:] - v_0[0:-1])/dx) - rho_0[0:-1]*v_0[0:-1]*\
    ((numpy.log(A[1:])-numpy.log(A[0:-1]))/dx) - v_0[0:-1]*((rho_0[1:]-rho_0[0:-1])/dx)
        
    v_f[:-1] = -v_0[0:-1] * ((v_0[1:] - v_0[0:-1])/dx) - (1./gamma)*(((T_0[1:] - T_0[0:-1])/dx)\
                                                                     + (T_0[0:-1]/rho_0[0:-1])*((rho_0[1:]-rho_0[0:-1])/dx))
        
    T_f[:-1] = -v_0[0:-1] * ((T_0[1:] - T_0[0:-1])/dx) - (gamma-1)*T_0[0:-1]*\
    ((v_0[1:] - v_0[0:-1])/dx+ v_0[0:-1]*((numpy.log(A[1:])-numpy.log(A[0:-1]))/dx))
        
    #rho_f[-1] = -rho_0[-2]*((v_0[-1] - v_0[-2])/dx) - rho_0[-2]*v_0[-2]*\
    #((numpy.log(A[-1])-numpy.log(A[-2]))/dx)-v_0[-2]*((rho_0[-1]-rho_0[-2])/dx)
        
    #v_f[-1] = -v_0[-2] * ((v_0[-1] - v_0[-2])/dx) - (1./gamma)*(((T_0[-1] - T_0[-2])/dx)\
     #                                                           + (T_0[-2]/rho_0[-2])*((rho_0[-1]-rho_0[-2])/dx))
        
    #T_f[-1] = -v_0[-2] * ((T_0[-1] - T_0[-2])/dx) - (gamma-1)*T_0[-2]*\
    #((v_0[-1] - v_0[-2])/dx +v_0[-2]*((numpy.log(A[-1])-numpy.log(A[-2]))/dx))
        
    # 2
        
    rho_g[0:-1] = rho_0[0:-1] + rho_f[0:-1]*dt
    v_g[0:-1] = v_0[0:-1] + v_f[0:-1]*dt
    T_g[0:-1] = T_0[0:-1] + T_f[0:-1]*dt
        
    # 3
        
    rho_h = -rho_g[1:-1]*((v_g[1:-1] - v_g[0:-2])/dx) -\
    rho_g[1:-1]*v_g[1:-1]*((numpy.log(A[1:-1])-numpy.log(A[0:-2]))/dx)-v_g[1:-1]*((rho_g[1:-1]-rho_g[0:-2])/dx)
        
    v_h = -v_g[1:-1] * ((v_g[1:-1] - v_g[0:-2])/dx) -\
    (1./gamma)*(((T_g[1:-1] - T_g[0:-2])/dx) + (T_g[1:-1]/rho_g[1:-1])*((rho_g[1:-1]-rho_g[0:-2])/dx))
        
    T_h = -v_g[1:-1] * ((T_g[1:-1] - T_g[0:-2])/dx) -\
    (gamma-1)*T_g[1:-1]*((v_g[1:-1] - v_g[0:-2])/dx +v_g[1:-1]*((numpy.log(A[1:-1])-numpy.log(A[0:-2]))/dx))
        
    #rho_h[0] = -rho_g[1]*((v_g[1] - v_g[0])/dx) - rho_g[1]*v_g[1]*\
    #((numpy.log(A[1])-numpy.log(A[0]))/dx)-v_g[1]*((rho_g[1]-rho_g[0])/dx)
        
    #v_h[0] = -v_g[1] * ((v_g[1] - v_g[0])/dx) - (1./gamma)*\
    #(((T_g[1] - T_g[0])/dx) + (T_g[1]/rho_g[1])*((rho_g[1]-rho_g[0])/dx))
        
    #T_h[0] = -v_g[1] * ((T_g[1] - T_g[0])/dx) - (gamma-1)*T_g[1]\
    #*((v_g[1] - v_g[0])/dx +v_g[1]*((numpy.log(A[1])-numpy.log(A[0]))/dx))
        
    # 4
        
    rho_j = 0.5*(rho_h + rho_f[1:-1])
    v_j = 0.5*(v_h + v_f[1:-1])
    T_j = 0.5*(T_h + T_f[1:-1])
     
    # 5
        
    rho_n[1:-1] = rho_0[1:-1] + rho_j*dt
    v_n[1:-1] = v_0[1:-1] + v_j*dt
    T_n[1:-1] = T_0[1:-1] + T_j*dt
    
    # Boundary Conditions
    
    rho_n[0] = 1.0
    v_n[0] = 2*v_n[1] - v_n[2]
    T_n[0] = 1.0
        
    rho_n[-1] = 2*rho_n[-2] - rho_n[-3]
    v_n[-1] = 2*v_n[-2] - v_n[-3]
    T_n[-1] = 2*T_n[-2] - T_n[-3]
        
    # Pressure and Mach Number
    P = rho_n*T_n
    Ma = v_n/(T_n**0.5)
    
    rho[t] = rho_n[15]
    T[t] = T_n[15]
    Ma_1[t] = Ma[15]
    P_1[t] = P[15]
```

```python
pyplot.plot(ntd,rho,'b-',linewidth=1)
pyplot.xlabel(r'$Number\, of\,time\, steps$')
pyplot.ylabel(r'$\rho/\rho_0$')   
pyplot.title('Quasi 1D Nozzle Flows'+' (Density)')
pyplot.show()

pyplot.plot(ntd,T,'b-',linewidth=1)
pyplot.xlabel(r'$Number\, of\, time\, steps$')
pyplot.ylabel(r'$T/T_0$')   
pyplot.title('Quasi 1D Nozzle Flows'+' (Temp.)')
pyplot.show()

pyplot.plot(ntd,P_1,'b-',linewidth=1)
pyplot.xlabel(r'$Number\, of\, time\, steps$')
pyplot.ylabel(r'$p/p_0$')   
pyplot.title('Quasi 1D Nozzle Flows'+' (Pressure)')
pyplot.show()

pyplot.plot(ntd,Ma_1,'b-',linewidth=1)
pyplot.xlabel(r'$Number\, of\, time\, steps$')
pyplot.ylabel(r'$Ma$')   
pyplot.title('Quasi 1D Nozzle Flows'+' (Mach)')
pyplot.show()
```
