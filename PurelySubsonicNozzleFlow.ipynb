```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
nx = numpy.linspace(0,3,31)
dx = 0.1
dt = 0.0267
gamma = 1.4
nt = 1400
ntd = numpy.linspace(0,nt,nt)

rho = numpy.zeros(31)
T = numpy.zeros(31)

# initial conditions
A = 1 + 2.2*(nx-1.5)**2

rho[:6] = 1.
T[:6] = 1.

rho[6:16] = 1.0 - 0.366*(nx[6:16]-0.5)
T[6:16] = 1.0 - 0.167*(nx[6:16]-0.5)

rho[16:] = 0.634 - 0.3879*(nx[16:]-1.5)
T[16:] = 0.833 - 0.3507*(nx[16:]-1.5)

V = 0.59/(rho*A)
p = rho*T

m = numpy.ones(31)*0.59

U_1 = rho*A
U_2 = rho*A*V
U_3 = rho*(T/(gamma-1) + gamma/2*V**2)*A

rho_n = numpy.zeros(nt)
T_n = numpy.zeros(nt)

#F_1 = U_2
#F_2 = U_2**2/U_1 + (gamma-1)/gamma*(U_3 - gamma/2*U_2**2/U_1)
#F_3 = gamma*U_2*U_3/U_1 - gamma*(gamma-1)/2*U_2**3/U_1**2
#J_2 = 1/gamma*rho[0:-1]*T[0:-1]*(A[1:]-A[0:-1])/dx
```

```python
for t in range(nt):
    #Predictor
    #Step 1
    F_1 = U_2
    F_2 = U_2**2/U_1 + (gamma-1)/gamma*(U_3 - gamma/2*U_2**2/U_1)
    F_3 = gamma*U_2*U_3/U_1 - gamma*(gamma-1)/2*U_2**3/U_1**2
    J_2 = 1/gamma*rho[0:-1]*T[0:-1]*(A[1:]-A[0:-1])/dx
    
    dU_1 = -(F_1[1:]-F_1[0:-1])/dx
    dU_2 = -(F_2[1:]-F_2[0:-1])/dx+J_2
    dU_3 = -(F_3[1:]-F_3[0:-1])/dx
    
    pU_1 = U_1[0:-1] + dU_1*dt
    pU_2 = U_2[0:-1] + dU_2*dt
    pU_3 = U_3[0:-1] + dU_3*dt
    
    prho = pU_1/A[0:-1]
    pT = (gamma-1)*(pU_3/pU_1 - gamma/2*(pU_2/pU_1)**2)
    
    pF_1 = pU_2
    pF_2 = pU_2**2/pU_1 + (gamma-1)/gamma*(pU_3 - gamma/2*pU_2**2/pU_1)
    pF_3 = gamma*pU_2*pU_3/pU_1 - gamma*(gamma-1)/2*pU_2**3/pU_1**2
    
    #Corrector
    cU_1 = -(pF_1[1:]-pF_1[0:-1])/dx
    cU_2 = -(pF_2[1:]-pF_2[0:-1])/dx + 1/gamma*prho[0:-1]*pT[0:-1]*(A[1:-1]-A[0:-2])/dx
    cU_3 = -(pF_3[1:]-pF_3[0:-1])/dx
    
    cdU_1 = 0.5*(dU_1[1:] + cU_1)
    cdU_2 = 0.5*(dU_2[1:] + cU_2)
    cdU_3 = 0.5*(dU_3[1:] + cU_3)
    
    U_1[1:-1] = U_1[1:-1] + cdU_1*dt
    U_2[1:-1] = U_2[1:-1] + cdU_2*dt
    U_3[1:-1] = U_3[1:-1] + cdU_3*dt
    
    U_1[0] = A[0]
    U_2[0] = 2*U_2[1] - U_2[2]
    V[0] = U_2[0]/U_1[0]
    U_3[0] = U_1[0]*(T[0]/(gamma-1)+gamma/2.0*V[0]**2)
    
    U_1[-1] = 2*U_1[-2] - U_1[-3]
    U_2[-1] = 2*U_2[-2] - U_2[-3]
    U_3[-1] = 2*U_3[-2] - U_3[-3]
 
    rho_n[t] = rho[15]
    T_n[t] = T[15]
    
    rho = U_1/A
    V = U_2/U_1
    T = (gamma-1)*(U_3/U_1 - gamma/2*(V)**2)
```

```python
pyplot.plot(ntd,rho_n,'b-',linewidth=1)
pyplot.xlabel(r'$Number\, of\,time\, steps$')
pyplot.ylabel(r'$\rho/\rho_0$')   
pyplot.title('(Density)')
pyplot.show()

pyplot.plot(ntd,T_n,'b-',linewidth=1)
pyplot.xlabel(r'$Number\, of\, time\, steps$')
pyplot.ylabel(r'$T/T_0$')   
pyplot.title('(Temperature)')
pyplot.show()
```
